```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.action_chains import ActionChains

# Automatically manage ChromeDriver version with webdriver-manager (for Google Chrome)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Open the PrizePicks website
driver.get('https://www.prizepicks.com/')

# Set up an explicit wait for elements to load
wait = WebDriverWait(driver, 10)  # Increase wait time to ensure elements load properly

# Wait until the "Log In" button is clickable and click it
login_button = wait.until(EC.element_to_be_clickable((By.ID, 'navbar-login-button')))
ActionChains(driver).move_to_element(login_button).click().perform()  # Click using ActionChains for robustness

# Debugging: print status
print("Clicked 'Log In' button.")

# Optional: Pause execution and let the user handle the CAPTCHA manually if needed
input("Complete the CAPTCHA and press Enter to continue...")

# Step 1: Use CSS Selector to locate and enter email
email_input = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'input#email-input')))
email_input.click()  # Click the email input field

# Use JavaScript to set the email value (if .send_keys() is not working)
driver.execute_script("arguments[0].value = '';", email_input)  # Replace with your actual email

# Debugging: print status
print("Entered email using JavaScript.")

# Step 2: Use CSS Selector to locate and enter password
password_input = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'input[type="password"]')))
password_input.click()  # Click the password input field

# Use JavaScript to set the password value (if .send_keys() is not working)
driver.execute_script("arguments[0].value = '';", password_input)  # Replace with your actual password

# Debugging: print status
print("Entered password using JavaScript.")

# Step 3: Locate and click the "Sign In" button using ActionChains to ensure it's clicked properly
sign_in_button = wait.until(EC.element_to_be_clickable((By.ID, 'submit-btn')))
ActionChains(driver).move_to_element(sign_in_button).click().perform()

# Debugging: print status
print("Clicked 'Sign In' button.")

# Wait for the page to load after signing in
wait.until(EC.url_changes("https://app.prizepicks.com/board"))

# Debugging: print status
print("Logged in successfully, waiting on logged-in page.")

# Pause to inspect the logged-in page
input("Press Enter to close the browser after signing in...")

# Close the browser
driver.quit()
```

# by extracting element

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.action_chains import ActionChains

# Automatically manage ChromeDriver version with webdriver-manager
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Open the PrizePicks website
driver.get('https://www.prizepicks.com/')

# Set up an explicit wait for elements to load
wait = WebDriverWait(driver, 30)  # Increased wait time to ensure dynamic elements are loaded

# Debugging: print status
print("Opened PrizePicks website.")

# Wait until the "Log In" button is clickable and click it using ActionChains
login_button = wait.until(EC.element_to_be_clickable((By.ID, 'navbar-login-button')))
ActionChains(driver).move_to_element(login_button).click().perform()

# Debugging: print status
print("Clicked 'Log In' button.")

# Optional: Pause execution to allow CAPTCHA handling if needed
input("Complete the CAPTCHA and press Enter to continue...")

# Debugging: print status
print("Waiting for email input field...")

# Step 1: Locate the email input field using the provided ID
try:
    email_input = wait.until(EC.presence_of_element_located((By.ID, 'email-input')))
    email_input.click()  # Click the email input field
    email_input.send_keys("")  # Replace with your actual email
    print("Email entered successfully.")
except Exception as e:
    print(f"Error locating or entering email: {e}")

# Debugging: print status
print("Waiting for password input field...")

# Step 2: Locate the password input field using the type and placeholder attributes
try:
    password_input = wait.until(EC.presence_of_element_located((By.XPATH, "//input[@type='password' and @placeholder='Type here...']")))
    password_input.click()  # Click the password input field
    password_input.send_keys("")  # Replace with your actual password
    print("Password entered successfully.")
except Exception as e:
    print(f"Error locating or entering password: {e}")

# Step 3: Locate and click the "Sign In" button using ActionChains to ensure it's clicked properly
try:
    sign_in_button = wait.until(EC.element_to_be_clickable((By.ID, 'submit-btn')))
    ActionChains(driver).move_to_element(sign_in_button).click().perform()
    print("Clicked 'Sign In' button.")
except Exception as e:
    print(f"Error clicking 'Sign In' button: {e}")

# Wait for the page to load after signing in
try:
    wait.until(EC.url_changes("https://app.prizepicks.com/login"))
    print("Logged in successfully, waiting on logged-in page.")
except Exception as e:
    print(f"Error during login process: {e}")

# Pause to inspect the logged-in page
input("Press Enter to close the browser after signing in...")

# Close the browser
driver.quit()
```
