```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

# Path to your WebDriver executable
PATH = "chromedriver.exe"

# Your login credentials
username = ''
password = ""

# URL of the login page
url = "https://app.kenect.com"

# Set up the WebDriver
service = Service(PATH)
driver = webdriver.Chrome(service=service)

# Initialize an empty dictionary to store the data
data_dict = {}

try:
    # Open the login page
    driver.get(url)

    # Wait for the email/username input field to be present
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "username"))
    )

    # Locate the email/username input field and enter the username
    email_input = driver.find_element(By.ID, "username")
    email_input.send_keys(username)

    # Locate the password input field and enter the password
    password_input = driver.find_element(By.ID, "password")
    password_input.send_keys(password)

    # Wait for the login button to be enabled
    login_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//button[@data-cy='loginSubmitButton']"))
    )

    # Click the login button
    login_button.click()

    # Pause for 5 seconds to observe the page
    time.sleep(5)

    # Wait for the icon to be present
    icon_button = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//span[contains(@class, 'kui-icon icon-contact standard')]"))
    )

    # Click the icon button
    icon_button.click()

    # Pause for 5 seconds to allow page load
    time.sleep(5)

    # Get the contact list container
    contact_list_container = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//div[@data-testid='contact-list-container']"))
    )

    # Initialize an empty set to store seen contacts
    seen_contacts = set()
    contact_count = 0
    end_of_list = False

    while not end_of_list:
        # Get all contact buttons currently visible
        contact_buttons = driver.find_elements(By.XPATH, "//a[contains(@class, 'imH1LJ22XOn5OIiaHD5G N0E7d2wCcMCdHBEWA946')]")

        new_contacts_found = False

        for button in contact_buttons:
            person_name = button.text

            if person_name not in seen_contacts:
                seen_contacts.add(person_name)
                new_contacts_found = True
                contact_count += 1

                # Split the person's name into first name and last name
                first_name, last_name = person_name.split(' ', 1) if ' ' in person_name else (person_name, '')

                # Click the person's button using JavaScript to avoid ElementClickInterceptedException
                driver.execute_script("arguments[0].click();", button)

                # Wait for 2 seconds to observe the action
                time.sleep(2)

                # Retrieve the SMS Outbound Messages status
                try:
                    sms_status = driver.find_element(By.XPATH, "//div[@title='Outbound Messages']//span[contains(text(), 'Text Allowed') or contains(text(), 'Do Not Text')]").text
                except:
                    sms_status = "Status not found"

                # Store the data in the dictionary
                data_dict[person_name] = {'First Name': first_name, 'Last Name': last_name, 'SMS Outbound': sms_status}

                # Print the person's name and SMS Outbound Messages status
                print(f"{first_name} {last_name}: {sms_status}")

                # Pause for 2 seconds before moving to the next person
                time.sleep(2)

                # Scroll the contact list container down to load more contacts after every 100 contacts
                if contact_count % 100 == 0:
                    driver.execute_script("arguments[0].scrollBy(0, arguments[0].scrollHeight);", contact_list_container)
                    time.sleep(5)  # Add a slight delay to allow for scrolling and loading

        if not new_contacts_found:
            # Perform an extra scroll if no new contacts were found
            driver.execute_script("arguments[0].scrollBy(0, arguments[0].scrollHeight);", contact_list_container)
            time.sleep(5)  # Add a slight delay to allow for scrolling and loading

            # Check for new contacts after the extra scroll
            contact_buttons = driver.find_elements(By.XPATH, "//a[contains(@class, 'imH1LJ22XOn5OIiaHD5G N0E7d2wCcMCdHBEWA946')]")
            for button in contact_buttons:
                person_name = button.text
                if person_name not in seen_contacts:
                    seen_contacts.add(person_name)
                    new_contacts_found = True
                    contact_count += 1

                    # Split the person's name into first name and last name
                    first_name, last_name = person_name.split(' ', 1) if ' ' in person_name else (person_name, '')

                    # Click the person's button using JavaScript to avoid ElementClickInterceptedException
                    driver.execute_script("arguments[0].click();", button)

                    # Wait for 2 seconds to observe the action
                    time.sleep(2)

                    # Retrieve the SMS Outbound Messages status
                    try:
                        sms_status = driver.find_element(By.XPATH, "//div[@title='Outbound Messages']//span[contains(text(), 'Text Allowed') or contains(text(), 'Do Not Text')]").text
                    except:
                        sms_status = "Status not found"

                    # Store the data in the dictionary
                    data_dict[person_name] = {'First Name': first_name, 'Last Name': last_name, 'SMS Outbound': sms_status}

                    # Print the person's name and SMS Outbound Messages status
                    print(f"{first_name} {last_name}: {sms_status}")

                    # Pause for 2 seconds before moving to the next person
                    time.sleep(2)

            if not new_contacts_found:
                end_of_list = True

finally:
    # Close the WebDriver
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

# Path to your WebDriver executable
PATH = "chromedriver.exe"

# Your login credentials
username = ''
password = ""

# URL of the login page
url = "https://app.kenect.com"

# Set up the WebDriver
service = Service(PATH)
driver = webdriver.Chrome(service=service)

# Initialize an empty dictionary to store the data
data_dict = {}

try:
    # Open the login page
    driver.get(url)

    # Wait for the email/username input field to be present
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "username"))
    )

    # Locate the email/username input field and enter the username
    email_input = driver.find_element(By.ID, "username")
    email_input.send_keys(username)

    # Locate the password input field and enter the password
    password_input = driver.find_element(By.ID, "password")
    password_input.send_keys(password)

    # Wait for the login button to be enabled
    login_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//button[@data-cy='loginSubmitButton']"))
    )

    # Click the login button
    login_button.click()

    # Pause for 5 seconds to observe the page
    time.sleep(5)

    # Wait for the icon to be present
    icon_button = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//span[contains(@class, 'kui-icon icon-contact standard')]"))
    )

    # Click the icon button
    icon_button.click()

    # Pause for 5 seconds to allow page load
    time.sleep(5)

    # Get the contact list container
    contact_list_container = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//div[@data-testid='contact-list-container']"))
    )

    # Scroll to the very bottom of the contact list to load all contacts
    while True:
        initial_height = driver.execute_script("return arguments[0].scrollHeight;", contact_list_container)
        driver.execute_script("arguments[0].scrollTo(0, arguments[0].scrollHeight);", contact_list_container)
        time.sleep(2)  # Wait for new contacts to load
        new_height = driver.execute_script("return arguments[0].scrollHeight;", contact_list_container)
        if new_height == initial_height:
            break

    # Initialize an empty set to store seen contacts
    seen_contacts = set()

    # Process all the contacts
    contact_buttons = driver.find_elements(By.XPATH, "//a[contains(@class, 'imH1LJ22XOn5OIiaHD5G N0E7d2wCcMCdHBEWA946')]")

    for button in contact_buttons:
        person_name = button.text

        if person_name not in seen_contacts:
            seen_contacts.add(person_name)

            # Split the person's name into first name and last name
            first_name, last_name = person_name.split(' ', 1) if ' ' in person_name else (person_name, '')

            # Click the person's button using JavaScript to avoid ElementClickInterceptedException
            driver.execute_script("arguments[0].click();", button)

            # Wait for 2 seconds to observe the action
            time.sleep(2)

            # Retrieve the SMS Outbound Messages status
            try:
                sms_status = driver.find_element(By.XPATH, "//div[@title='Outbound Messages']//span[contains(text(), 'Text Allowed') or contains(text(), 'Do Not Text')]").text
            except:
                sms_status = "Status not found"

            # Store the data in the dictionary
            data_dict[person_name] = {'First Name': first_name, 'Last Name': last_name, 'SMS Outbound': sms_status}

            # Print the person's name and SMS Outbound Messages status
            print(f"{first_name} {last_name}: {sms_status}")

            # Pause for 2 seconds before moving to the next person
            time.sleep(2)

finally:
    # Close the WebDriver
    driver.quit()
```

```python
df = pd.DataFrame.from_dict(data_dict, orient='index').reset_index()
df.columns = ['Full Name', 'First Name', 'Last Name', 'SMS Outbound']
df.drop('Full Name', axis=1, inplace=True)
df.head()
```
