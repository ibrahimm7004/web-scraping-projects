```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Hardcoded Facebook credentials
FB_EMAIL = 'ibrahim.mahmood7@hotmail.com'
FB_PASSWORD = 'sh3637Gdhj'

def login_to_facebook_and_redirect(driver, group_url):
    # Open Facebook login page
    driver.get('https://www.facebook.com/login')
    
    # Enter email
    driver.find_element(By.ID, "email").send_keys(FB_EMAIL)
    
    # Enter password
    driver.find_element(By.ID, "pass").send_keys(FB_PASSWORD)
    
    # Click the login button
    driver.find_element(By.NAME, "login").click()

    # Wait for login to complete by sleeping for a few seconds
    time.sleep(5)
    
    # Redirect to the group URL after login
    driver.get(group_url)
    
    print("Redirected to the Facebook group.")

def handle_notification_overlay(driver):
    try:
        # Wait for the notification overlay to appear and ensure it is clickable
        block_button = WebDriverWait(driver, 15).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(., 'Block')]"))
        )
        # Click the "Block" button
        block_button.click()
        print("Blocked notifications.")
        time.sleep(1)  # Give some time for the overlay to disappear

    except Exception as e:
        print("No notification overlay found or another issue occurred:", e)

def click_members_link(driver):
    try:
        # Wait for the members link to appear and be clickable
        members_link = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, "a[href$='/members/']"))
        )

        # Scroll the element into view
        driver.execute_script("arguments[0].scrollIntoView(true);", members_link)
        time.sleep(1)  # Give some time for scrolling

        # Use JavaScript to click the element
        driver.execute_script("arguments[0].click();", members_link)

        print("Clicked on the 'members' link.")

    except Exception as e:
        print(f"An error occurred: {e}")

def get_members_page_html(driver):
    try:
        # Wait for the members page to load
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "div"))
        )

        # Get the page source (HTML)
        page_html = driver.page_source
        print("Members page HTML has been retrieved.")
        return page_html

    except Exception as e:
        print(f"An error occurred while trying to retrieve the members page HTML: {e}")

def visit_public_group_and_get_members_html(link):
    service = Service('chromedriver.exe')
    driver = webdriver.Chrome(service=service)
    
    try:
        # Log in to Facebook and redirect to the group
        login_to_facebook_and_redirect(driver, link)

        # Add a delay to ensure the group page loads completely
        time.sleep(3)

        # Handle notification overlay if it appears
        #handle_notification_overlay(driver)
        
        # Click on the "members" link
        click_members_link(driver)

        # Add a delay to ensure the members page loads completely
        time.sleep(3)

        # Get the members page HTML
        members_html = get_members_page_html(driver)

        # Return the HTML code
        return members_html

    finally:
        # The browser can remain open if needed; otherwise, uncomment the line below to close it
        # driver.quit()
        pass

# Example usage
link = 'https://www.facebook.com/groups/309526382084446/'
members_html = visit_public_group_and_get_members_html(link)
```

```python
from bs4 import BeautifulSoup

# Assuming members_html is already defined and contains the HTML code
def extract_urls_from_html(members_html):
    # Parse the HTML with BeautifulSoup
    soup = BeautifulSoup(members_html, 'html.parser')
    
    # Find all 'a' tags with an href attribute
    links = soup.find_all('a', href=True)
    
    # Extract the href attribute from each 'a' tag
    urls = [link['href'] for link in links]
    
    # Return the list of URLs
    return urls

# Example usage
# members_html = ... # your HTML content here
urls = extract_urls_from_html(members_html)

# Print or process the URLs
for url in urls:
    print(url)
```

```python
def extract_member_urls(urls):
    # Define the base path that identifies member URLs
    member_url_pattern = "/groups/309526382084446/user/"
    
    # Filter the URLs that contain the member URL pattern
    member_urls = [url for url in urls if member_url_pattern in url]
    
    return member_urls

# Example usage
# The 'urls' list should be the result from the previous extraction
member_urls = extract_member_urls(urls)

# Print or process the member URLs
for url in member_urls:
    print(url)
```

```python
def convert_to_proper_url(broken_url):
    # Extract the numeric ID from the broken URL
    numeric_id = broken_url.split('/')[-2]  # Get the second last segment of the URL
    
    # Construct the proper Facebook profile URL
    proper_url = f"https://www.facebook.com/profile.php?id={numeric_id}"
    
    return proper_url
```

```python
list_set = []
for i in member_urls:
    list_set.append(convert_to_proper_url(i))

new_set = set(list_set)

print(new_set)
```
