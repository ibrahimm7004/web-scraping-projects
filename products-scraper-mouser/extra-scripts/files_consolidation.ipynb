### Retrieve File Names

```python
import os
import pandas as pd

# Define the folder path
folder_path = r'HomePage'

# Check if the folder exists
if not os.path.exists(folder_path):
    print(f"Folder {folder_path} does not exist.")
    exit(1)

# Get the list of all files in the folder
file_names = os.listdir(folder_path)

# Create a pandas DataFrame with the file names
df_files = pd.DataFrame(file_names, columns=["File Name"])
```

```python
file_path = r'files_done.tsv
df_files.to_csv(file_path, sep='\t', index=False, header=False)
```

### Consolidate Files

```python
import os
import shutil

# Define the directories
dir1 = r'DetailPage'
dir2 = r'DetailPage'

# Ensure the target directory exists
if not os.path.exists(dir1):
    print(f"Target directory {dir1} does not exist.")
    exit(1)

# Iterate through all the files in dir2
for file_name in os.listdir(dir2):
    # Check if the file is an HTML file
    if file_name.endswith(".html"):
        # Construct full file paths
        source_file = os.path.join(dir2, file_name)
        destination_file = os.path.join(dir1, file_name)

        # Check if the file already exists in dir1
        if os.path.exists(destination_file):
            print(f"File {file_name} already exists in {dir1}, skipping.")
        else:
            # Move the file
            shutil.move(source_file, destination_file)
            print(f"Moved {file_name} to {dir1}")

print("File transfer completed.")
```
