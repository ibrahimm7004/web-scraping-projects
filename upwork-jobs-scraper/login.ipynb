```python
import time
import random
import json
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
from dotenv import load_dotenv
import os

# Load environment variables from .env
load_dotenv()

# Load Upwork credentials from the .env file
UPWORK_EMAIL = os.getenv('UPWORK_EMAIL')
UPWORK_PASSWORD = os.getenv('UPWORK_PASSWORD')

# Driver setup with realistic headers and User-Agent
driver_path = 'chromedriver.exe'
chrome_options = Options()
chrome_options.add_argument("--start-maximized")

# Use a realistic User-Agent to avoid detection
chrome_options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari/537.36"
)

# Disguise the fact that we are using an automated browser
chrome_options.add_argument("--disable-blink-features=AutomationControlled")
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
chrome_options.add_experimental_option('useAutomationExtension', False)

service = Service(executable_path=driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# Function to simulate random delays
def random_delay(min_delay=2, max_delay=5):
    time.sleep(random.uniform(min_delay, max_delay))

# Function to simulate page scrolling
def simulate_scrolling():
    scroll_pause_time = random.uniform(2, 4)  # Random pause time between scrolling actions
    scroll_height = driver.execute_script("return document.body.scrollHeight")
    
    for i in range(0, scroll_height, random.randint(200, 400)):
        driver.execute_script(f"window.scrollTo(0, {i});")
        random_delay(scroll_pause_time, scroll_pause_time + 1)

# Function to simulate mouse movement
def simulate_mouse_movement():
    action = ActionChains(driver)
    # Randomly move mouse around the page
    for i in range(5):
        x_offset = random.randint(10, 50)
        y_offset = random.randint(10, 50)
        action.move_by_offset(x_offset, y_offset).perform()
        random_delay(0.5, 1.5)

# Login URL
login_url = "https://www.upwork.com/ab/account-security/login?redir=%2Fnx%2Ffind-work%2F"

# Open the Upwork login page
driver.get(login_url)
random_delay(3, 6)

# Simulate page scrolling and mouse movement
simulate_scrolling()
simulate_mouse_movement()

# Load cookies from file (assuming cookies are saved in 'upwork_cookies.json')
with open('upwork_cookies.json', 'r') as file:
    cookies = json.load(file)

# Add cookies to the browser
for cookie in cookies:
    if 'expiry' in cookie:
        del cookie['expiry']  # Remove expiry field if present
    if 'sameSite' in cookie and cookie['sameSite'] not in ["Strict", "Lax", "None"]:
        del cookie['sameSite']  # Fix sameSite attribute
    try:
        driver.add_cookie(cookie)
    except Exception as e:
        print(f"Error adding cookie: {cookie['name']}, Error: {e}")

# Refresh the page after adding cookies
driver.refresh()
random_delay(3, 6)

# Check if login is required, and login if necessary
try:
    # If cookies are valid, you might already be logged in after refresh.
    # Check if you're still on the login page
    current_url = driver.current_url
    if "login" in current_url:
        print("Cookies are not sufficient, proceeding with login.")

        # Enter email
        email_input = driver.find_element(By.ID, "login_username")
        email_input.send_keys(UPWORK_EMAIL)

        # Click "Continue with Email"
        driver.find_element(By.XPATH, "//button[@data-cy='login-continue-with-email']").click()
        random_delay(2, 4)

        # Enter password
        password_input = driver.find_element(By.ID, "login_password")
        password_input.send_keys(UPWORK_PASSWORD)

        # Click "Sign In"
        driver.find_element(By.XPATH, "//button[@data-cy='login-button']").click()

        random_delay(5, 8)  # Wait for login to complete
    else:
        print("Logged in using cookies!")
except Exception as e:
    print(f"Login error: {e}")

# Simulate some more scrolling and mouse movement to continue mimicking human behavior
simulate_scrolling()
simulate_mouse_movement()

# Close the browser after a few seconds
random_delay(5, 10)
driver.quit()
```
